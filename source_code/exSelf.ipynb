```python
#Request는 url을 가지고 서버에 요청하는 역할
#urlopen은 request한 내용을 받아서 url로 여는 역할
from urllib.request import urlopen, Request 
import urllib
```

```python
# url이 ../여명의_눈동자 
# 인데 깨져서 나올 때
# 1) url decode 구글에 검색해서 디코딩 후에 복붙해도 되고,
# 2) {search_words}로 넣은뒤에 request할 때 format해서 넣어도 된다.
html = "https://ko.wikipedia.org/wiki/{search_words}"
# 여기서는 2)방법 사용
# Request하는데 url의 포맷에서 {search_words}부분을 "여명의_눈동자"를 인코딩한 내용을 삽입
req = Request(html.format(search_words=urllib.parse.quote("여명의_눈동자")))
response = urlopen(req) #응답받은 내용을 url로 열어서 대입
response, response.status #response 값은 무슨 뜻? ## status가 2XX면 정상응답
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response, "html.parser")
```

```python
n=0 
for each in soup.find_all("ul"): #ul 태그를 쭉 살펴보면서 
    print("=>"+str(n)+"="*10) #몇번째 ul태그에 등장인물이 있는지 확인...
    print(each.get_text()) # 여기서는 16번째(인덱스 15)에 있음
    n+=1
```

## 시카고 맛집 데이터 50개 리스트 가져오기

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
# 왜 잘라서 한다고 했더라..
url_base = "https://www.chicagomag.com/"
url_sub = "Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/"
url = url_base + url_sub
```

```python
import ssl
from fake_useragent import UserAgent #클라이언트를 랜덤으로 생성해주는 패키지
ua = UserAgent()
context = ssl._create_unverified_context() #ssl 해결방법
req = Request(url, headers={"User-agent": "Chrome"}) ## 403 Error 해결방법 1)
# 정석은 관리자모드에서 Network 맨밑에 client 값 가져와야하고 "Chrome"은 약식 해결 방법
req = Request(url, headers={"User-agent": ua.ie}) ## 403 Error 해결방법 2)
response = urlopen(req,context=context) ## ssl 및 403 해결 후 응답
response.status
```

```python
html = urlopen(req,context=context)
soup = BeautifulSoup(html, "html.parser")
#soup.prettify()
```

```python
#div 태그에 class가 sammy인 곳에 각 랭킹 정보가 있음
soup.find("div", class_="sammy") #랭킹 1위 정보
len(soup.findAll("div", class_="sammy")) # 총 50개 있음을 확인
```

```python
sample = soup.find("div",class_="sammy") # sample로 랭킹 하나를 가져와 봤음
```

```python
import re # 정규식(regular expression) 패키지
sample_string = sample.find(class_="sammyListing").get_text() # class가 sammyListing인 태그에서 url_sub 내용이 담겨있다.
# [상동] sample.select_one("div.sammyListing").text
re.split(("\n|\r\n"),sample_string) # \n 또는 \r\n으로 seperate한 결과를 리스트에 넣는다.
```

```python
# 필요한 내용을 담을 리스트를 선언
url_base = "https://www.chicagomag.com/"
rank = []
main_menu  = []
cafe_name = []
url_add = []
# 리스트에 하나씩 컬럼을 생성한 후에 DataFrame으로 합친다.
list_soup = soup.find_all("div","sammy") #랭킹에 있는 50개 태그를 모두 가져온다.
# 랭킹 1에 접근하기 위해서는 list_soup[0]
from urllib.parse import urljoin
for each_soup in list_soup:
    rank.append(each_soup.find("div",class_="sammyRank").text )
    tmp_string = each_soup.find("div",class_="sammyListing").text
    main_menu.append(re.split(("\n|\r\n"),tmp_string)[0])
    cafe_name.append(re.split(("\n|\r\n"),tmp_string)[1])
    url_add.append(urljoin( url_base, each_soup.find("a")["href"]))
list_soup[0]

```

```python
import pandas as pd
# 데이터 프레임 생성 방법 1)
data = {
    "Rank" : rank,
    "Menu" : main_menu,
    "Cafe" : cafe_name,
    "URL" : url_add
}
df = pd.DataFrame(data)
# 데이터 프레임 생성 방법 2)
# 컬럼을 Rank, Cafe, Menu, URL 순서로 변경
#df = pd.DataFrame(data, columns = ["Rank","Cafe","Menu","URL"])
df.head()
```

```python
# 가격 / 주소 가져오기
df = pd.read_csv("../data/03. best_sandwiches_list_chicage.csv", index_col=0)
# df["URL"]에 들어가서 가격과 주소 태그 가져오기
df["URL"][0]
```

```python
req = Request(df["URL"][0], headers={"user-agent":ua.ie})
html = urlopen(req,context=context).read() #read가 있고 없고는 무슨차이?
soup_tmp = BeautifulSoup(html, "html.parser")
soup_tmp.find("p", class_="addy")
```

```python
# 정규표현식
price_tmp = soup_tmp.find("p",class_="addy").get_text()
price_tmp = re.split((",."),price_tmp)[0]
price_tmp # 가격과 주소가 같이 있음
price = re.search("\$\d+\.(\d+)?", price_tmp) # 정규표현식으로 서칭
price = re.search("\$\d+\.(\d+)?", price_tmp).group() #에서 matcing된 값만
address = price_tmp[len(price)+2:] # +2는 ". "를 생략
price, address
```

```python
#위에 내용 합치면
## 거기에 진행정도 출력하는 
from tqdm import tqdm
price = []
address = []
for idx,row in tqdm(df.iterrows()):
    req = Request(row["URL"], headers={"user-agent":ua.ie})
    html = urlopen(req,context=context).read()
    soup_tmp = BeautifulSoup(html, "html.parser")
    ####
    gettings = soup_tmp.find("p", class_="addy").get_text()
    price_tmp = re.split(".,",gettings)[0]
    tmp = re.search("\$\d+\.(\d+)?", price_tmp).group()
    price.append(tmp) #가격데이터 추가
    address.append(price_tmp[len(tmp)+2 : ]) #2는 띄어쓰기
```

```python
df["Price"] = price
df["Address"] = address
df.head()
```

```python
import numpy as np

lat = []
lng = []

for idx, row in tqdm(df.iterrows()):
    if not row["Address"] == "Multiple location":
        target_name = row["Address"] + "," + "Chicago"
        #print(target_name)
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get("geometry")
        lat.append(location_output["location"]["lat"])
        lng.append(location_output["location"]["lng"])
    else :
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
# 자동화 코드
## 날짜변경
#url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20210930"
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date={date}"
date = pd.date_range("2021.03.01", periods = 100, freq="D") #freq : 하루기준으로 100일 만듬
date[0]
```

```python
import time
from tqdm import tqdm

movie_date = []
movie_name = []
movie_point = []

for today in tqdm(date): #tqdm은 얼마나 진행되는지 보여주는 것 #date는 2021-01-01 ~2021-04-30 까지
    url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date={date}"
    response = urlopen(url.format(date=today.strftime("%Y%m%d"))) #strftime : 날짜형태
    soup = BeautifulSoup(response, "html.parser")

    end = len(soup.findAll("td","point"))

    movie_date.extend([today for _ in range(0, end)])
    movie_name.extend([soup.select("div.tit5")[n].find("a").text for n in range(0,end)])
    movie_point.extend([soup.find_all("td",class_="point")[n].string for n in range(0, end)])

    time.sleep(0.5) 
```
