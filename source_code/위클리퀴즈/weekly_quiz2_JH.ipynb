```python
df2 = subway_passengers_df.pivot_table(index=["노선명"], values=["승차총승객수","하차총승객수"], aggfunc=np.sum)
df2
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
from matplotlib import rc 

# plt.rcParams["axes.unicode_minus"] = False 
rc("font", family="Malgun Gothic") # Windows:  
# %matplotlib inline 
get_ipython().run_line_magic("matplotlib", "inline")
```

```python
df6 = df5[["승차총승객수","사용일자"]].set_index("사용일자")
df6.loc[:,'승차총승객수'] = df6['승차총승객수'].astype('int')

plt.figure(figsize=(10, 6))
plt.plot(df6.index, df6['승차총승객수'])
plt.title("강남역 승차총승객수 21년 8월")
plt.show()
```

```python
df7 = subway_passengers_df.loc[subway_passengers_df["역명"]=="팔당"]
df7.loc[:,"사용일자"] = df7.loc[:,"사용일자"].astype('str')
df7.loc[:,"사용일자"] =  pd.to_datetime(df7.loc[:,'사용일자'])
df7.loc[:,'승차총승객수'] = df7['승차총승객수'].astype('int')
df7
```

```python
df7 = df7[["승차총승객수","사용일자"]].set_index("사용일자")

plt.figure(figsize=(10, 6))
plt.plot(df7.index, df6['승차총승객수'], df7['승차총승객수']*10)
plt.title("강남역과 팔당역 승차총승객수 21년 8월")
plt.legend(["강남역", "팔당역*10"])
plt.show()
```
