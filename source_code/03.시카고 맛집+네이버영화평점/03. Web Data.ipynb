```python
#최상위 bs4 폴더에서 BeautifulSoup만 import
from bs4 import BeautifulSoup
```

```python
# html 열기
page = open("../data/03. test_first.html").read()

#그냥 page써도 되지만 print쓰면 가독성 좋게 출력
#print(page)

# page라는 객체를 html.parser 엔진으로 읽어옴
## https://www.crummy.com/software/BeautifulSoup/bs4/doc/
soup = BeautifulSoup(page, "html.parser")
#print(soup) ##대신 prettify 메서드로 보면 더 가독성 좋음
print(soup.prettify()) 
```

## 2) find_all()
 - 여러 개의 tag를 반환
 - 리스트 형태로 반환

```python
#특정 태그 확인 -> 텍스트만 출력
soup.find_all(class_="outer-text")[0].text
soup.find_all(class_="outer-text")[1].string
#soup.find_all(class_="outer-text")[1].get_text()
```

```python
# p 태그 리스트에서 텍스트 속성만 출력
for each_tag in soup.find_all("p"):
    print("="*50)
    print(each_tag.text,end="")
```

```python
# a 태그에서 href 속성값에 있는 값 추출
links = soup.findAll("a")
## href 값만 출력  
links[0].get("href") # 1) get 메서드
links[1]['href']     # 2) 리스트 요소 가져오는 형태
```

```python

for each in links :
    href = each.get("href")
    text = each.get_text()
    print( text + " ======(link)=====> " + href)
```

## BeautifulSoup 예제 1-2 - 네이버 금융
 - !pip install requests
 - find, find_all
 - 과 유사한 기능
 - select, select_one
 - find, select_one : 단일 선택
 - find_all, select : 다중 선택

```python
#2개 기능 거의 유사... 둘 중 편한걸 사용하면됨
import requests
#from urllib.request.Request

from bs4 import BeautifulSoup
```

```python
url = "https://finance.naver.com/marketindex/"
response = requests.get(url)
#requests.get()
#requests.post() 방식 2개가 있음
#response -> <Response [200]>
soup = BeautifulSoup(response.text, "html.parser") 
#이미 response 객체에 담겨있는 web 내용을 가독성있게 보기 위해 bs 모듈 사용
print(soup.prettify())
'''여기까지는 1-1과 과정 동일'''
```

```python
#soup.find_all("li", class_="on") 와 동일
exchangeList = soup.select("#exchangeList > li", ) #class => . , id => # 사용
#len(exchangeList) -> 4
exchangeList
```

```python
#추출할 데이터 5개
#find도 가능하지만 select_one 코드가 간결...
title = exchangeList[0].select_one(".h_lst").text
# <h3 class="h_lst"><span class="blind">미국 USD</span></h3>
exchange = exchangeList[0].find(class_="value").text
change = exchangeList[0].select_one(".change").text
# div 태그에 head~up 클래스 바로 아래(>) blind 클래스인 것을 가져옴 
updown = exchangeList[0].select_one("div.head_info > .blind").text
'''div가 밑에는 왜 없어졌지?????'''
title, exchange, change, updown
```

```python
baseUrl = "https://finance.naver.com"
#위 4개와 달리 a 태그에서 href 속성값을 get함
link = baseUrl + exchangeList[0].select_one("a").get("href")
link
```

---

```python
# 위의 4개 데이터 수집
import pandas as pd

exchange_datas = []
baseUrl = "https://finance.naver.com"

for item in exchangeList :
    data = { #딕셔너리타입
        "title" : item.select_one(".h_lst").text,
        "exchange" : item.find(class_="value").text,
        "change" : item.select_one(".change").text,
        "updown" : item.select_one(".head_info > .blind").text,
        "link" : baseUrl + item.select_one("a").get("href")
    }
    exchange_datas.append(data)
#data
df = pd.DataFrame(exchange_datas)
df.to_excel("./naver.finanace.xlsx",encoding="utf-8")
```
